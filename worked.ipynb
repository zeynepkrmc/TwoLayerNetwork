```python
df = pd.read_csv('data.csv')
```

Training the model

```python
prediction_vars = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
                   'radius_se','area_se', 'radius_worst', 'compactness_worst']
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.15, random_state=1)
```

```python
train_x = train[prediction_vars]
train_y = train['diagnosis']
test_x = test[prediction_vars]
test_y = test['diagnosis']
```

```python
train_x
```

```python
test_y
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model = MLPClassifier() #define model
```

```python
#random forest
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
test_y
```

```python
#how to improve, asses, confussion matrix?
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predictions)
```

Burda bir matris gösterdi malignant ve benign olmasıyla ilgili. 4 kare top left malignant olan ve malig. tespit edilen, top right.., bottom left.., bottom right....

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision = precision_score(test_y, predictions)
print("The prediction is %.2f",precision)

recall = recall_score(test_y, predictions)
print("The recall is %.2f",recall)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test_y, predictions)
print("The accuracy is %.2f",accuracy)
#we have  initial model till here
```
